```python
import pandas as pd
import openpyxl as op

"""Retrieving the forgotten Password"""


def forgotPassword():
    print("Enter the 'USERNAME'")
    value = 0
    val = input()
    df2 = pd.read_excel('D:/python/reg_login/reg_login.xlsx')
    for i in range(df2[df2.columns[0]].count()):
        if(val == df2.iloc[i,0]):
            value = i
            break
    retrivedForPassword = df2.iloc[value,1]
    return retrivedForPassword
    

"""checking weather the username is existed or not during Login"""

def checkLoginUserName(loginUserName):
    count = 0
    list1 = [1]
    list2 = [0]
    str = loginUserName
    df1 = pd.read_excel('D:/python/reg_login/reg_login.xlsx')
    for i in range(df1[df1.columns[0]].count()):
        if(str == df1.iloc[i,0]):
            list1.append(i)
            break
        else :
            count+=1
    if (count == df1[df1.columns[0]].count()) :

        return list2
    else :
        return list1

"""checking weather the username is already existed or not during registration"""
    
def check(username):
    str = username
    df1 = pd.read_excel('D:/python/reg_login/reg_login.xlsx')
    for i in range(df1[df1.columns[0]].count()):
        if(str == df1.iloc[i,0]):
            return 1
            

"""Username validation"""

def validateUserName(username):
    listSpecialCharacters = ['!','@','#','$','%','^','&','*','?']
    listNumbers = ['1','2','3','4','5','6','7','8','9','0']
    countAtTheRate = 0
    countDot = 0
    countSpecial = 0
    countNumbers = 0
    str = username        
    m=0
    while m<len(listNumbers):
        if(str[0] == listNumbers[m]):
            countNumbers+=1
            break
        m+=1
    n=0
    while n<len(listSpecialCharacters):
        if(str[0] == listSpecialCharacters[n]):
            countSpecial+=1
            break
        n+=1
    if(countSpecial == 1 or countNumbers == 1):
        print("The 'USERNAME' canot start with enther 'SPECIAL CHARACTER' or 'NUMBERS' ")
        print("Please enter the 'USERNAME' correctly")
        val = input()
        username = validateUserName(val)
    i = 0
    while i<len(str)-1:
        if(str[i] == "@" and str[i+1] == "."):
            print("'.' is immediately following '@'")
            print("Please enter the 'USERNAME' correctly")
            val = input()
            username = validateUserName(val)
        i+=1
    j=0
    while j<len(str):
        if(str[j] == "@"):
            countAtTheRate+=1
            k =j
            while k<len(str):
                if(str[k] == "."):
                    countDot+=1
                k+=1
        j+=1
    if(countAtTheRate == 0 or countDot == 0):
        print("Either '@' or '.' was not found in the 'USERNAME'")
        print("Please enter the 'USERNAME' correctly")
        val = input()
        username = validateUserName(val)
    if(countAtTheRate >=1 and countDot >1):
        print("The frequency of '@' or the frequency of '.' after the '@' are more")
        print("Please enter the 'USERNAME' correctly")
        val = input()
        username = validateUserName(val)
    #print(userName)
    return username

"""password validation"""

def validatePassword(password):
    str = password
    countSpecial = 0
    countNumbers = 0
    countUpper = 0
    countLower = 0
    listSpecialCharacters = ['!','@','#','$','%','^','&','*','?']
    listNumbers = ['1','2','3','4','5','6','7','8','9','0']
    listUpperCase = ['Q','W','E','R','T','Y','U','I','O','P','A','S','D','F','G','H','J','K','L','Z','X','C','V','B','N','M']
    listLowerCase = ['q','w','e','r','t','y','u','i','o','p','a','s','d','f','g','h','j','k','l','z','x','c','v','b','n','m']
    if len(str)<5 or len(str)>15:
        print("The 'PASSWORD' length should be in a range of '5' to '16'")
        print("Enter the 'PASSWORD' correctly ->")
        val = input()
        password = validatePassword(val)
    i=0
    while i<len(listSpecialCharacters):
        j=0
        while j<len(str):
            if(str[j] == listSpecialCharacters[i]) :
                countSpecial+=1
            j+=1
        i+=1
    i=0
    while i<len(listNumbers):
        j=0
        while j<len(str):
            if(str[j] == listNumbers[i]) :
                countNumbers+=1
            j+=1
        i+=1
    i=0
    while i<len(listUpperCase):
        j=0
        while j<len(str):
            if(str[j] == listUpperCase[i]) :
                countUpper+=1
            j+=1
        i+=1
    i=0
    while i<len(listLowerCase):
        j=0
        while j<len(str):
            if(str[j] == listLowerCase[i]) :
                countLower+=1
            j+=1
        i+=1
    #print(countSpecial)
    #print(countNumbers)
    #print(countUpper)
    #print(countLower)
    if(countSpecial == 0 or countNumbers == 0 or countUpper == 0 or countLower == 0) :
        print(
            '''The 'PASSWORD' Must have minimum one special character,
              one digit,
              one uppercase, 
              one lowercase character
             '''
             )
        print("Enter the 'PASSWORD' correctly ->")
        val = input()
        password = validatePassword(val)
    return password

"""Main Function"""

def regLoginFunction():
    print("Enter which functionality do you need to proceed 'REGISTER' or 'LOGIN' Write in lowercase ")
    functionality = input()
    if(functionality == "register"):
        print("Enter the 'USERNAME'")
        userName = input()
        valUserName = validateUserName(userName)
        print("Enter the 'PASSWORD'")
        password = input()
        valPassword = validatePassword(password)
        checkUser = check(userName)
        if(checkUser == 1):
            print("The entered 'USERNAME' is already existed")
            print("Please choose the 'LOGIN' functionality")
            return 0
        
        """Storing the Details in to the file"""
        
        df = pd.DataFrame({'Username':[valUserName],'Password':[valPassword]})
        with pd.ExcelWriter("D:/python/reg_login/reg_login.xlsx",mode="a",engine="openpyxl",if_sheet_exists="overlay") as writer:
            df.to_excel(writer, sheet_name="Sheet1",header=None, startrow=writer.sheets["Sheet1"].max_row,index=False)
            
        print("Successfully Registered")
        
    elif(functionality == "login"):
        print("Enter the 'Username' -> ")
        loginUserName = input()
        cheLoginUserName = checkLoginUserName(loginUserName)
        if(cheLoginUserName[0] == 0):
            print("The entered 'USERNAME' does not exist ")
            print("Plese select the 'REGISTER' functionality")
            return 0
        print("Enter the 'PASSWORD' -> ")
        loginPassword = input()
        df2 = pd.read_excel('D:/python/reg_login/reg_login.xlsx')
        retrivedPassword = df2.iloc[cheLoginUserName[1],1]
        if(loginPassword == retrivedPassword):
            print("Successful Login")
        else :
            print('''
                    The entered 'PASSWORD' does not match
                    enter the 'FORGOT' option to retrive the original password in lowercase
            ''')
            value1 = input()
            if(value1 == 'forgot') :
                retrivedForgotPassword = forgotPassword()
                print("Successful Retrival")
                print(retrivedForgotPassword)
           
    else :
        print("Enter the functionality correctly")
        regLoginFunction()
regLoginFunction()
```
